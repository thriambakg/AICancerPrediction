## Parse

```python
import pandas as pd

df = pd.read_excel("Pathology_Project/Student_Version.xlsx")

patient_data = []

for index, row in df.iterrows():
  try:
    sample = int(row["sample"])
    data = pd.read_csv(f"Pathology_Project/713_1217Spectra_csv_Tranformed/sample{sample}.csv", header=None, names=["Column1", "Column2"])
    data = data["Column2"].tolist()
    patient_data.append(tuple([row['Survival']] + data))
  except Exception as e:
    pass
```

```python
df = pd.DataFrame(patient_data)
df.head()
```

```python
df.to_csv('parsed_patient_data.csv', index=False) # might need to change this
```
