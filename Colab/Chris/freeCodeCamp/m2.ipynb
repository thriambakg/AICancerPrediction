Creating variables (which are tensors). This one has rank 0

```python
my_tensor_string = tf.Variable("I am Chris", tf.string)
print(my_tensor_string)
```

Tensors are multidimensional vectors. Their rank is how many dimensions they have.

The first example is one, storing three elements.
The second is two, storing two elements

```python
rank_1 = tf.Variable([5, 6, 7], tf.int16)
rank_2 = tf.Variable([[1, 2]], tf.int16)
print(tf.rank(rank_1))
print(tf.rank(rank_2))
```

# Shape

The shape of a tensor is how many elements are in each of its layers

```python
print(rank_1.shape)
print(rank_2.shape)
```

# Reshape

You can change the shape of a tensor using the `.reshape` method.  
As long as the numbers multiply to be identical, then it will work

```python
tensor1 = tf.ones([1, 2, 3])
print(tensor1)
tensor2 = tf.reshape(tensor1, [3, 2, 1])
print(tensor2)
tensor3 = tf.reshape(tensor2, [3, -1])
print(tensor3)
```
