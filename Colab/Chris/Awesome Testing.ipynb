## Import Pandas

```python
import pandas as pd
```

## Read Main File

```python
df = pd.read_excel("Pathology_Project/Student_Version.xlsx")
df.head()
```

```python
survival = df["Survival"]
survival.head()
```

## Read the Valid Samples

```python
valid_y = []
valid_X = []

for index, sample in df["sample"].items():
  try:
    data = pd.read_csv(f"Pathology_Project/713_1217Spectra_csv_Tranformed/sample{sample}.csv", header=None, names=["Column1", "Column2"])
    valid_X.append(data["Column2"])
    valid_y.append(survival[index])
  except Exception as e:
    pass
```

## Visualize Data

```python
import tensorflow as tf
import matplotlib.pyplot as plt

def visualize(index):
  print(valid_y[index])
  record = tf.convert_to_tensor(valid_X[index])
  plt.plot(record)

visualize(3)
```

## Create `X` and `y` from Data

### Import Tensorflow

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

# Parse Valid Data

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(valid_X, valid_y, test_size=0.2)
```

## Convert to Tensors

```python
X_train = tf.constant(X_train, dtype=tf.float16)
X_test = tf.constant(X_test, dtype=tf.float16)
y_train = tf.constant(y_train, dtype=tf.float16)
y_test = tf.constant(y_test, dtype=tf.float16)
```

# Creating Model

```python
# from sklearn.model_selection import train_test_split
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tf.random.set_seed(50)

model = tf.keras.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(1000),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

history = model.fit(X_train, y_train, epochs=100)

pd.DataFrame(history.history).plot()
plt.ylabel("Loss Function")
plt.xlabel("Epochs")
```

```python
model.evaluate(X_test, y_test)
```
