```python
import pandas as pd

df = pd.read_csv("parsed_patient_data.csv")
df.head()
```

```python
y = df["0"]
y.head()
```

```python
X = df.drop("0", axis=1)
X.head()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

def visualize(index):
  print(y[index])
  plt.plot(X.loc[index].tolist())

visualize(7)
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler = scaler.fit(X_train)

X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
X_train = tf.constant(X_train, dtype=tf.float32)
X_test = tf.constant(X_test, dtype=tf.float32)
y_train = tf.constant(y_train, dtype=tf.float32)
y_test = tf.constant(y_test, dtype=tf.float32)
```

models

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tf.random.set_seed(50)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1000),
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

history_1 = model_1.fit(X_train_norm, y_train, epochs=100)
```

```python
pd.DataFrame(history_1.history).plot()
plt.ylabel("Loss Function")
plt.xlabel("Epochs")
model_1.evaluate(X_test_norm, y_test)
```

```python
tf.random.set_seed(50)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1000),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["mae"])
history_2 = model_2.fit(X_train_norm,
                        y_train,
                        epochs=100)
```

```python
pd.DataFrame(history_2.history).plot()
plt.ylabel("Loss Function")
plt.xlabel("Epochs")
model_2.evaluate(X_test_norm, y_test)
```

```python
tf.random.set_seed(50)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(1000),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss='mse',
                optimizer='adam',
                metrics=['mae'])

history_3 = model_3.fit(X_train_norm, y_train, epochs=100)
```

```python
pd.DataFrame(history_3.history).plot()
plt.ylabel("Loss Function")
plt.xlabel("Epochs")
model_3.evaluate(X_test_norm, y_test)
```

```python
tf.random.set_seed(50)

model_4 = tf.keras.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(1000),
    tf.keras.layers.Dense(1)
])

model_4.compile(loss='mse',
                optimizer='adam',
                metrics=['mae'])

history_4 = model_4.fit(X_train, y_train, epochs=25)
```

```python
pd.DataFrame(history_4.history).plot()
plt.ylabel("Loss Function")
plt.xlabel("Epochs")
model_4.evaluate(X_test, y_test)
```
